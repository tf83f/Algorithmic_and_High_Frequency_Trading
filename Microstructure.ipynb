# Microstructure

# Microprice
Microstructure focus on liquidity and price discovery which are two primary aspects that drive the questions and issues behind the design of effective and high frequency trading.

Micro-price (weighted mid-price) incorporates information on order imbalance. Relatively larger quantity of offers on the bid than on the ask indicates greater buying pressure, and the ‘true price’ is closer to the ask than to the bid.

$S_{t}^{*} = \frac{V_{t}^{b}}{V_{t}^{b} + V_{t}^{a}} P_{t}{a} + \frac{V_{t}^{a}}{V_{t}^{b} + V_{t}^{a}} P_{t}{b}$

# Interarrival times 
However smaller (than 1 second) intervals require discrete model to study the micro-price. We can't keep looking at asset return and have to look at interarrival time of change. When we look at ACF of interarrivals time bid and ask are not independent but have a string autocorrelated component. It is an empirical fact that they cluster. Rapid changes in the bid/ask are followed by further rapid changes, while a relatively long calm period is also similarly followed by another calm period. Like asset returns, inter-arrival times can be modeled by a power law distribution with heavy tails 

$P(x_{i} \le x) = 1 - (\frac{k}{x})^{\alpha}$

We also note that big $ stock, means that they can experience much smaller % changes in their prices because of the minimum tick size and would thus experience more price changes.

Since volume and volatility are highly correlated and display strong time series persistence, any variable correlated with volatility will, inevitably, possess nontrivial forecast power for future volatility. This is true for bid-ask spreads, the quote intensity, the transaction count, the (normalized) trading volume.

Even though prices changes tend to be reversed, the direction of the current price changes (whether it is an uptick or downtick) does not carry additional information about future prices changes.

# Intraday activity 

Explain the peaks at the beginning and the end of the day. The common hypothesis about the factors that drive volume:
- New information generates greater volume. But this does not explain the magnitude of the increase in volume at the end of the day.
- Larger volume at the end of the day: traders who have not been able to meet their liquidation target will accelerate trade execution as the market approaches its close. A second not unrelated possible explanation is that traders may prefer to postpone non urgent executions towards the end of the day when execution price are lower (i.e the impact of orders (walking the LOB) is lower at the end of the trading day. Finally, strategies that target volume will naturally exacerbate the increase in volume anticipated during that period.

We used two basic arguments to explain intraday patterns, that new information increases trading volume, and that an increased desire to trade (e.g. due to increase trader urgency) interacts with the quality of the market which feeds back to motivate further trading. Market quality enters directly into the second argument: an expected increase in volume generates expectations of better market quality, that is of improved effectiveness of the market in facilitating trade via lower execution costs (spreads), greater price efficiency (less mean-reversion in price levels or lower transitory volatility), which induces grater volume.

# Spreads

Spreads measure the immediate cost of executing a trade aggressively.

Mid-price $S_{t} = \frac{1}{2}(a_{t}+b_{t}$ is interpreted as the true price (equilibrium price), but the true price may in reality be closer to the bid or the ask, and hence sometimes it might be wiser to use micro-price. 

The spread may arise for different reasons (market frictions, compensation for inventory risk, or adverse selection, …). 

The two most common spread measure :
- Quoted spread (QS) is the difference between the ask and the bid prices $QS_{t} = a_{t} – b_{t}$ and represents the potential cost of immediacy. The direct trading cost of a market sell order would be $S_{t} – b_{t} = \frac{QS_{t}}{2}$ while that of a market buy would be $ a_{t}  - S_{t}= \frac{QS_{t}}{2}$
- In contrast, the effective (half) spread, ES measures the realized difference between the price paid and the mid-price, buy order $ES_{t} = a_{t}  - S_{t}$ and for sell $ES_{t} = S_{t} - b_{t}  $. For a MO against a visible LO it will be the quoted half spread, greater if it walk the LOB or even negative against an hidden order in the spread, hence $ES \le QS$. 
- Also ES can only be measured when there is a trade, while quoted spreads are always observable. Therefore, it may be possible that quoted spreads differ from effective spreads if market conditions around traders systematically different from those without trades. 
- ES is less volatile that QS, this would happen if executions tend to be concentrated around times of narrows quotes spreads.
- The afternoon is associated with increase trading, hence we find as hypothesized, that during a period with a constant flow of information more trading and lower spreads occur together. This connection between trading volume and spreads fails during the morning where the situation is completely reversed. Declining volumes go hand-in-hand with declining spreads. This can be explained by appealing to other factor affecting volume, like information. When the market opens, the market absorb all information since the close, this would explain the heavier trading. But a lot of new is also associated with a great deal of uncertainty, which leads to wider spread. 

# Volatility 
Volatility measure the effectiveness of the price in transmitting information about the market value of an asset.

The literature differentiates between fundamental volatility (fluctuation in the true market price) and microstructure noise (due to the way market operates). The two combined is named the raw volatility

The one cent tick size is a binding constraint for small stock price. This leads to an unusual degree of price stickiness as many small price movements are not sufficient to push the bid or ask a whole cent away from their currents price. Thus despite having similar activity as another price, it will display lower volatility. 

Comparing the intraday volatility pattern to the intraday volume, we see a common pattern, high at the beginning of the day, lower as the day pass until it reaches a place, followed by an increase until market close. 

# Market depth and trade size

Depth measures the amount of visible liquidity in the market 

Market quality is not just about the informational content of prices or the cost of executing a small order, it is also about depth (the volume posted on the LOB available for immediate execution).

Depth and trade size are not independent of one another: the decision of how much to trade depends on the expected availability of shares resting as LOs for immediate execution and similarly, the decision of how much to offer will depend on the expected flow of incoming MOs.

If the depth is thin, MOs will be small, which implies that in thin markets, relatively urgent large orders that would walk the LOB need to be broke, up into smaller MOs which are then sequentially executed over a period of time. 

As expected there is the usual sharp increase at the end of the trading day, this is consistent with greater market quality in the form of narrower spreads and the increased desire to close positions. Somewhat surprising is that depth is also higher at the beginning of the day, where we hypothesized that high price uncertainty leads to wider spreads. Thus suggest that the theoretical tradeoff between the benefits of. Market marking from increased order arrival and the cost from higher uncertainty is resolved in practice in favor of providing greater liquidity.

# Price impact 

Trades can have permanent and temporary impact on prices

MOs do not arrive at random times, they tend to arrive when spread are narrow and opportunistically hit orders that are posted to the mid-price, so it is only natural that we should observe wider spread after an execution.  

Permanent price impact (b) is linear in the net order-flow. Permanent impact : impact of order flow on the change in midprice $\delta S_{n} = b \mu_{n} \epsilon_{n}$

Temporary price impact (k) is linear in the volume traded. Temporary price impact is the difference between the execution price that the trader receive and the best quote.

There is a positive relationship between b and k. Also there is the relationship between price impact and depth, so that days with little depth will be associated with high price impact. 

The largest price impact tends to occur in the morning, then this impact flattens and stays flat throughout the day and towards the end of the day it lessens.
